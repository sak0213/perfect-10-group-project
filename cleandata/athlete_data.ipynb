```python
#Dependencies
import pandas as pd
import numpy as np
athletes = pd.read_csv("clean_df")
```

```python
#fill null values with "n/a" and make series for the get_dummy function that pivots medals
athletes = athletes.fillna("n/a")
athletes = athletes[['ID', 'Name', 'Sex', 'Age', 'Height', 'Weight', 'Team','NOC', 'Games', 'Year', 'Season', 'City', 'Sport', 'Event', 'Medal']]
medals = pd.DataFrame({
    "Medal": athletes["Medal"]
}
)
```

```python
# make a df of the pivoted medals
medals = pd.get_dummies(medals)
medals.head()
```

```python
#ad pivoted medals and important columns into new main df
athletes = athletes[['ID', 'Name', 'Sex', 'Age', 'Height', 'Weight', 'Team','NOC', 'Games', 'Year', 'Season', 'City', 'Sport', 'Event','region_x', 'notes_x']]
athletes["Bronze"] = medals["Medal_Bronze"]
athletes["Gold"] = medals["Medal_Gold"]
athletes["Silver"] = medals["Medal_Silver"]
athletes["n/a"] = medals["Medal_n/a"]
athletes.head()

```

```python
#export a df of all important columns and pivoted medals
athletes.to_csv("athletes_and_medalcount")
```
